### Define the number of labels each feature can be categorised into

```python
nb_class = 5
labels = np.arange(1, nb_class + 1)
```

```python
df['R'] = (pd.cut(df['recency_7'], nb_class, labels=labels[::-1])).map(str)
df['F'] = (pd.cut(df['frequency_7'], nb_class, labels=labels)).map(str)
df['M'] = (pd.cut(df['monetary_value_7'], nb_class, labels=labels)).map(str)
```

```python
df['RFM_score'] = df.R.astype(str) + df.F.astype(str) + df.M.astype(str)
```

```python
sns.pairplot(df[['recency_7','frequency_7','monetary_value_7','F']], hue='F')
```

```python
sns.pairplot(df[['recency_7', 'frequency_7', 'monetary_value_7', 'M']], hue='M')
```

```python
df.groupby('RFM_score').id.count().sort_values(ascending=False)
```

```python
### This is a definition inspired from the internet, this can be changed based on the business definition. The main target of this work is to ensure we can decrease and increase customer categories to see its effect on the modeling side
```

```python
seg_map = {
    r'[1-2][1-2][1-5]': 'hibernating',
    r'[1-2][3-4][1-5]': 'at_Risk',
    r'[1-2]5[1-5]': 'cant_loose',
    r'3[1-2][1-5]': 'about_to_sleep',
    r'33[1-5]': 'need_attention',
    r'[3-4][4-5][1-5]': 'loyal_customers',
    r'41[1-5]': 'promising',
    r'51[1-5]': 'new_customers',
    r'[4-5][2-3][1-5]': 'potential_loyalists',
    r'5[4-5][1-5]': 'champions'
}

df["customer_category"] = df["RFM_score"].replace(seg_map, regex=True)
```

```python
df.groupby("customer_category").id.count().sort_values(ascending=False)
```

### The following categories of people across recency frequency and monetary helps one to take actions on where to focus

```python
rfm_cols = ['recency_7','frequency_7','monetary_value_7']
fig, ax = plt.subplots(3,1, figsize = (10,5))
for i,col in enumerate(rfm_cols):
    sns.barplot(df, x ='customer_category',y=col,ax=ax[i])
```
